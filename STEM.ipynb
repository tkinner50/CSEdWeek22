# Visualizing data with Python

## Have you ever wondered how scientists and engineers work with sets of data? 

While there are certainly still many times where professionals will plot data by hand or using a spreadsheet tool like Excel, there are many times where the amount of data gets too be large to handle with basic graphing tools. 

To explore the <i>why</i> and <i>how</i> of using Python to plot data, we'll use this Python notebook to open a large dataset. This dataset is obtained from a scientific study by the National Oceanic and Atmospheric Administration (NOAA) about the behavior of wild sea turtles. The sea turtles were outfitted with GPS-enabled devices that tracked various metrics about their behavior. 

We will use this dataset to explore the diving behavior of sea turtles to answer the question, <b>"How long do sea turtles stay underwater?"</b>. 

## Step 1 - Setting up our notebook

To get everything we will need, we are going to set up our environment. There are many steps to setting up a development environment, most of which are beyond what we will cover in our session. The majority of the development environment setup has been done for you, and now all we need to do is initialize our notebook by importing several libraries. 

In the Python langauge, libraries are lines of code that have been developed for use by others in the development community. Imagine you wanted to write a program that was able to send an email each time you pressed a button. Instead of writing every single element of that code by hand, we could likely find and use libraries to reduce our development time, including a library to help us send emails or recognize when a button has been pressed. 

Including libraries is a decision programmers make that weighs their need to develop code quickly and reliably, versus understanding and/or owning every line of code in a program. Much like how car manufacturers will buy their windshields from one supplier, and their car stereos from another supplier, we can use various libraries to create our program.
